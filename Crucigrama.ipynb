```python
crucigrama = [
    ['_', '_', '_', '_', '_', '_'],
    ['_', '_', '_', '_', '_', '_'],
    ['_', '_', '_', '_', '_', '_'],
    ['_', '_', '_', '_', '_', '_'],
    ['_', '_', '_', '_', '_', '_']
]

def obtener_entero_positivo(mensaje):
    while True:
        try:
            valor = int(input(mensaje))
            if valor >= 0:
                return valor
            else:
                print("Por favor, ingrese un valor entero positivo.")
        except ValueError:
            print("Por favor, ingrese un valor entero válido.")

def mostrar_crucigrama(crucigrama):
    print("  0 1 2 3 4 5")
    print(" +-+-+-+-+-+-")
    for i, fila in enumerate(crucigrama):
        print(f"{i}|{' '.join(fila)}|")
    print(" +-+-+-+-+-+-")
    print()

def llenar_palabra(crucigrama, palabra, fila, columna):
    for i in range(len(palabra)):
        crucigrama[fila][columna + i] = palabra[i]

print("¡Bienvenido al llenador de crucigramas!")
print("Ingresa palabras para llenar el crucigrama. Para salir, ingresa 'q'.\n")

mostrar_crucigrama(crucigrama)

while True:
    palabra = input("Ingresa una palabra (o 'q' para salir): ")

    if palabra == 'q':
        break

    fila = obtener_entero_positivo("Ingresa la fila donde quieres colocar la palabra: ")
    columna = obtener_entero_positivo("Ingresa la columna donde quieres colocar la palabra: ")

    llenar_palabra(crucigrama, palabra, fila, columna)
    mostrar_crucigrama(crucigrama)

print("Has salido del programa. Aquí está el crucigrama final:")
mostrar_crucigrama(crucigrama)
```
